### problem 2
Write a Python generator function called batch_generator. It should take a list of numbers (representing data samples) and a batch_size. The generator should yield one batch at a time as a list.

Example Input:
data = [1, 2, 3, 4, 5, 6, 7, 8]
batch_size = 3

Expected Output (when iterated):
[1, 2, 3]
[4, 5, 6]
[7, 8]


```python
# declaring the function with parameters
# for 0-10, it will be 0, 3, 6, 9. 
# for 0, it will go to the third line, will slice from 0, 0+3=3. so it will be 0, 1, 2. will continue the process with 3, 6, 9.

def batch_generator (data_samples, batch_size):
    for i in range (0, len(data_samples), batch_size):
        yield data_samples[i:i+batch_size]

# example data set
data = [x for x in range(10)]
size = 3

# data = [int(x) for x in input("give ur data").split(" ") ]
# size = int(input("size?"))

# will iterate through the process and print each time. 
for batch in batch_generator (data, size):
    print(batch)
```
